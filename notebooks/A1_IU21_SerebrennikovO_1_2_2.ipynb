<p>11. Определите как изменилось значение показателя в соответствии с Вашим вариантом с 2012 по 2020</p>
<p><b>Вариант 1.</b> На сколько позиций в рейтинге c 2012 по 2020</p>
<p><b>Вариант 2.</b> На сколько процентов значение CPI с 2012 по 2020</p>

```python
countries_cpi = df_asia[["Country"] + [f"cpi20{x}" for x in range(12,21)]].copy()
# countries_cpi["change"] = countries_cpi["cpi2012"] - countries_cpi["cpi2013"]

for year in range(2013, 2021):
    curr_cpi = "cpi{year}".format(year=year)    
    prev_cpi = "cpi{year}".format(year=year-1)
    countries_cpi["diff_of_" + curr_cpi + "-" + prev_cpi] = \
        (countries_cpi[curr_cpi] - countries_cpi[prev_cpi])/countries_cpi[prev_cpi]
for col in countries_cpi.columns:
    if col.startswith("diff"): 
        # firstly I added '%' symbol to each value with float to sting convertion, 
        # but it messed up the typing for future calculations of mean and other values
        
        countries_cpi[col] = pd.Series([round(val*100, 2) for val in countries_cpi[col]], index = countries_cpi.index)
countries_cpi

```

<p>12. Выведите </p>

```python
# here is the ugly plot for that.

plt.figure(figsize = (12, 12))
plt.title('Springfield')
plt.ylabel('CPI')
plt.yticks(range(-20, 50, 5))
label = ([f'20{i} Score' for i in range(12, 21)])
plt.xticks(np.arange(9), label)
plt.grid(axis='both', which='major') 

countries = countries_cpi['Country'].to_list()
final = list()

for country in countries:
    cpi = [countries_cpi.loc[countries_cpi.Country == country][f"diff_of_cpi{x}-cpi{x-1}"] for x in range(2013, 2021)]
    final.append([country, [cpi[x].values[0] for x in range(len(cpi))]])

import random
tmp = list()
for i in range(len(final)):
    
    plt.plot(final[i][1], color="gray",
             linewidth = 1,
             marker = 'o', markersize = 5)
    
    tmp.append(final[i][1])

    
# at the dataframe above you can see that we have a lot of NaN values in CPI scores, which eventually end up in NaN 
# diff for each year pair, so we forget about it there because if we don't - we need to fill NaN values with
# preceding values of scores... okay, but what if we don't have a preceding score like in Solomon Islands?
# this becomes more complicated. The solution is: leave the changes for this years equal to zero then we will
# not mess up the data and graphs will only move where the changes happen...
# проще говоря: нас не интересуют пустые значения так как мы не можем посчитать из них разницу между двумя годами
# поэтому мы просто заменяем разницу между такими годами равной нулю

no_nan_np = np.nan_to_num(np.array(tmp))

plt.plot(np.mean(no_nan_np, axis=0), color="blue",
             label="mean", linewidth = 3,
             marker = 'o', markersize = 7)
plt.plot(np.max(no_nan_np, axis=0), color="red",
             label="max", linewidth = 3,
             marker = 'o', markersize = 7)
plt.plot(np.min(no_nan_np, axis=0), color="yellow",
             label="min", linewidth = 3,
             marker = 'o', markersize = 7)
plt.legend(loc = 1, fontsize = 12)
```

<p>1. Загрузите данные в один dataframe из всех файлов в папке /data/stock. Все файлы имеют одинаковую структуру, в том числе наименование столбцов. В качестве значений индекса dataframe'а необходимо указать значения столбца "Date". Название столбцов должные соответствовать названию акций (имя файла без .csv), а их значения - значениям цены закрытия (столбец "Close" в файлах .csv)</p>

```python
import os
STOCK_FOLDER = "../data/A1_Descriptive_Analysis/stock/"
files = os.listdir(STOCK_FOLDER)
df_stocks = pd.DataFrame()
for file in files:
    stock = pd.read_csv(STOCK_FOLDER + str(file), index_col=0, parse_dates=True)
    df_stocks[file.split('.')[0]] = stock['Close']
df_stocks.head(5)
```
